```python
import pandas as pd
from sqlalchemy import create_engine
import psycopg2
```

# Import countries of the world csv

```python
csv_file = "countries of the world.csv"
cntryOfWorld_df = pd.read_csv(csv_file)
cntryOfWorld_df.head()
```

# Import World Happiness Report CSV

```python
csv_file = "world-happiness-report-2019.csv"
WorldHappy_df = pd.read_csv(csv_file)
WorldHappy_df.head()
```

# Clean data

```python
WorldHappy_df.stack().str.replace(',','.').unstack()
WorldHappy_df.head()
```

```python
WorldHappyClean_df = WorldHappy_df.rename(columns={ 
    "Country (region)":"country",
    "Ladder":"ladder",
    "SD of Ladder":"sdofladder",
    "Positive affect":"positiveaffect",
    "Negative affect":"negativeaffect",
    "Social support":"socialsupport",
    "Freedom":"freedom",
    "Corruption":"corruption",
    "Generosity":"generosity",
    "Log of GDP\nper capita":"logofgdppercapita",
    "Healthy life\nexpectancy":"lifeexpectancy"
})

WorldHappyClean_df.head()
```

```python
New_cntryOfWorld_df = cntryOfWorld_df[["Country","Region","Population","Area (sq. mi.)","Pop. Density (per sq. mi.)","Coastline (coast/area ratio)","Net migration","GDP ($ per capita)","Literacy (%)","Phones (per 1000)"]].copy()
New_cntryOfWorld_df.head()
```

```python
New_cntryOfWorld_df = New_cntryOfWorld_df.rename(columns={ 
    "Country":"country",
    "Region":"region",
    "Population":"population",
    "Area (sq. mi.)":"area",
    "Pop. Density (per sq. mi.)":"density",
    "Coastline (coast/area ratio)":"coastiness",
    "Net migration":"migration",
    "GDP ($ per capita)":"gdp",
    "Literacy (%)":"lituracy",
    "Phones (per 1000)":"phones"
})


New_cntryOfWorld_df.head()
```

```python
New_cntryOfWorld_df.stack().str.replace(',','.').unstack()

New_cntryOfWorld_df.head()
```

# Connect to local database

```python
rds_connection_string = "postgres:postgres@localhost:5432/happy_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

# insert data from df into database

```python
WorldHappyClean_df.to_sql(name='happy_table', con=engine, if_exists='append', index=True)
```

```python
pd.read_sql_query('select * from happy_table', con=engine).head()
```

```python
New_cntryOfWorld_df.to_sql(name='stats_table', con=engine, if_exists='append', index=True)
```

```python
pd.read_sql_query('select * from stats_table', con=engine).head()
```
